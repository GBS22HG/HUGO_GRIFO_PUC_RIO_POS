# Descrição do Problema

O conjunto de dados "Student Perfomance", é um conjunto de dados multivariado que contém as notas dos estudantes secundaristas de 2 escolas portuguesas ("Gabriel Pereira" e "Mousinho da Silveira") nas 3 avaliações de conteúdo da disciplina de Português, realizadas pelos estudantes  ao curso de um ano letivo (não informado o período do dataset).

O objetivo principal deste projeto, é correlacionar as informações pessoais e sócio-econômicas disponíveis no conjunto de dados, com o desempenho em avaliações de conteúdo dos estudantes, de modo a identificar características estatísticamente significativas no desempenho estudantil e, posteriormente, desenvolver um modelo preditivo para estimar as notas dos alunos.

# Hipóteses levantadas
Após uma análise das informações disponíveis, as hipóteses que serão estudadas neste projeto serão:

- **1 -** A construção familiar dos estudantes (tamanho da família, relacionamento com os pais, profissão dos familiares, etc..), gera impacto significante no resultado em avaliações escolares ?

-  **2 -** O tempo de estudo dedicado as disciplinas é um fator determinístico no resultado das avaliações? (Quanto maior o volume de horas estudadas, melhor é o desempenho na disciplina? Ou existe um ponto de saturação?)
  
- **3 -** Comportamentos socialmente questionáveis (como ausências escolares, reprovações, consumo de álcool em dias úteis, etc...) impactam negativamente o resultado das avaliações?

- **4 -** Sabemos que o resultado final possui alta correlação com a nota das avaliaçãos. Contudo, conseguimos prever o resultado final na disciplina de um estudante, sem que ele tenha realizado nenhuma avaliação a partir dos demais atributos conhecidos?

## Tipo de Problema

Este é um problema de **regressão supervisionada**. Dado um conjunto de características (sociais, econômicas, discplinares), o objetivo é prever qual será a nota dos estudantes ao final do ano letivo.

**Observação**: O conjunto de dados escolhido não possui informações sobre a aprovação dos estudantes (exemplo: Informação Binária - 0, "não foi aprovado" ou 1, "foi aprovado") e não sabemos a nota de "corte" das disciplinas, apenas os resultados nas 3 avaliações. Portanto, para evitar viéses de classificação, não iremos definir uma nota de "corte" arbitrária e e não iremos realizar testes de **classificação supervisonada**.

*exemplo de viés:* estudantes que obtiveram uma nota próxima a nota de corte na 1ª avaliação podem aumentar o ritmo de estudos com a esperança de serem aprovados, enquanto estudantes que tiveram um péssimo desempenho, na mesma avaliação, podem desistir de serem aprovados e não se empenharem nas disciplinas.

#Média
A média é uma medida de tendência central que representa o valor típico ou o ponto de equilíbrio de um conjunto de dados. É calculada somando-se todos os valores e dividindo-se pelo número total de observações. É sensível a valores extremos (outliers).

```python
# Criando uma função para plotar gráficos de barra pela média de qualquer variavel:
def plot_media_g3_por_variavel(ax, grupo, titulo, variavel, xlabel,ylabel):
    media = df.groupby(grupo)[variavel].mean() #função que calcula a média agrupada
    media_total = df[variavel].mean() # função que calcula a média geral
    sns.barplot(x=media.index, y=media.values, ax=ax) #função de plot de gráfico barra
    ax.set_title(titulo) #Título Gráfico
    ax.set_xlabel(xlabel) #Título Eixo X
    ax.set_ylabel(ylabel) #Título Eixo Y
    ax.set_ylim(0,20) #Limite eixo y

# Linha horizontal indicando a média geral
    ax.axhline(media_total, color='red', linestyle='--', linewidth=1.5, label=f'Média das notas: {media_total:.2f}')
    ax.legend(loc='upper right')

# Adicionando rótulo de dados
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.2f}',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points')
```

```python
# Criar dois subplots lado a lado
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 15))
axes = axes.flatten()

# Plotar a Média das Notas Finais (G3) por Variáveis Familiares:
plot_media_g3_por_variavel(axes[0], 'Mjob', 'Nota Média por Profissão da Mãe','G3','Profissão','Nota Média')
plot_media_g3_por_variavel(axes[1], 'Fjob', 'Nota Média por Profissão do Pai','G3','Profissão','Nota Média')
plot_media_g3_por_variavel(axes[2], 'Medu', 'Nota Média por Escolaridade da Mãe','G3','Escolaridade','Nota Média')
plot_media_g3_por_variavel(axes[3], 'Fedu', 'Nota Média por Escolaridade do Pai','G3','Escolaridade','Nota Média')
plot_media_g3_por_variavel(axes[4], 'famsize', 'Nota Média pelo tamanho da família','G3','Tamanho','Nota Média')
plot_media_g3_por_variavel(axes[5], 'famrel', 'Nota Média pelo Relacionamento Familiar','G3','Relacionamento','Nota Média')
```

A partir das **Médias bivariadas**, podemos concluir que:

#Histograma - (Bivariado)
A distribuição de dados descreve como os valores de uma variável se espalham, ou seja, a frequência com que diferentes valores ocorrem. Entender a distribuição é crucial na análise de dados, pois revela padrões, tendências centrais, dispersão e a presença de valores atípicos (outliers). O histograma é uma ferramenta visual fundamental para representar essa distribuição, mostrando a forma dos dados, se são simétricos ou assimétricos, unimodais ou multimodais.

```python
# Distribuição da nota dos alunos por tempo de estudo

#criando uma função que mostra o histograma por grupo

def plot_histograma_g3_por_variavel(df, grupo, variavel, bins=10, add_median=True, add_mean=True, kde=True):
    grupos = sorted(df[grupo].dropna().unique())  # garante ordem
    n = len(grupos) #quantidade de gráficos por linha

    # Plotando os histogramas lado a lado Definir layout dos subplots automaticamente (máx 3 por linha)

    fig, axes = plt.subplots(2, 2, figsize=(3 * n, 5 ), sharey=True)
    axes = axes.flatten()

    for i, valor in enumerate(grupos):
        subset = df[df[grupo] == valor][variavel].dropna()
        ax = axes[i]
        sns.histplot(subset, kde=False, bins=bins, ax=ax, color='skyblue')
        ax.set_title(f'Distribuição das Notas Finais (studytime = {valor})')
        ax.set_xlabel(variavel)
        ax.set_ylabel('Frequência')

        # Adiciona linha de média
        if add_mean:
            media = subset.mean()
            ax.axvline(media, color='red', linestyle='--', label=f'Média: {media:.2f}')

        # Adiciona linha de mediana
        if add_median:
            mediana = subset.median()
            ax.axvline(mediana, color='green', linestyle='-.', label=f'Mediana: {mediana:.2f}')

        ax.legend()

    plt.tight_layout()
    plt.show()
```

```python
# Distribuição da nota dos alunos por tempo de estudo
plot_histograma_g3_por_variavel(df, grupo='studytime', variavel='G3', bins=10)
```

A partir dos **Histogramas bivariados**, podemos concluir que:
- Uma quantidade muita pequena dos alunos estuda mais de 5 horas por semana (3, 4).
- A quantidade de horas semanais estudas possui forte relação com a nota final da discplina. Nenhum aluno que estudou mais de 5 horas recebeu uma nota 0.

#Matriz de Correlação
A matriz de correlação mede a força e a direção de uma relação linear que os atributos númericos podem ter. Valores próximos a 1 indicam uma forte correlação positiva, -1 uma forte correlação negativa, e 0 ausência de correlação linear.

```python
# Definindo uma função que calcula os coeficientes de pearson apenas paras as variáveis numéricas.
def matriz_correlacao_pearson(df, variavel_alvo='G3'):
    variaveis_numericas = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    if variavel_alvo not in variaveis_numericas:
        variaveis_numericas.append(variavel_alvo)

    matriz_pearson = df[variaveis_numericas].corr(method='pearson')

    print("\n=== Matriz de Correlação de Pearson (variáveis numéricas) ===")
    display(matriz_pearson.round(3))

    plt.figure(figsize=(10, 8))
    sns.heatmap(matriz_pearson, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
    plt.title("Matriz de Correlação de Pearson entre Variáveis Numéricas")
    plt.tight_layout()
    plt.show()

    return matriz_pearson
```

```python
# Plotando a matriz de Correlação de Pearson com as variáveis numéricas do df.
matriz_pearson = matriz_correlacao_pearson(df)
```

A partir da **Matriz de Correlação**, podemos concluir que:
- Teremos uma acurácia melhor no modelo de regressão supervisionada se utilizarmos as notas prévias dos alunos (G1 e G2) como atributos dependentes, pois existe uma forte correlação positiva com a nota final (G3).
- Os demais atributos, apresentam baixa correlação linear com a nota final do período, contudo, as variáveis apresentadas até o momento estão entre as mais explicativas.

# Codificação

A codificação dos dados é um processo importante em modelos de machine learning. Ele transforma variáveis categóricas em variáveis numéricas, permitindo a normalização e padronização dos dados.
Esse processo aumenta a quantidade de atributos no dataset, pois cria uma variável binária para todas os domínios do conjunto de dados original.



```python
# Aplica OneHotEncoder apenas nas colunas categóricas
colunas_categoricas = X_train.select_dtypes(include='object').columns

# Cria e aplica encoder
encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')

X_train_encoding = pd.DataFrame(encoder.fit_transform(X_train[colunas_categoricas]),
                               columns=encoder.get_feature_names_out(colunas_categoricas),
                               index=X_train.index)

X_test_encoding = pd.DataFrame(encoder.transform(X_test[colunas_categoricas]),
                              columns=encoder.get_feature_names_out(colunas_categoricas),
                              index=X_test.index)

# Junta com as colunas numéricas
X_train_num = X_train.drop(columns=colunas_categoricas).reset_index(drop=True)
X_test_num = X_test.drop(columns=colunas_categoricas).reset_index(drop=True)

X_train_encoded = pd.concat([X_train_num, X_train_encoding.reset_index(drop=True)], axis=1)
X_test_encoded = pd.concat([X_test_num, X_test_encoding.reset_index(drop=True)], axis=1)
```
